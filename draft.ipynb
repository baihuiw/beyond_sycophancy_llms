```python
import pandas as pd

# Load the data
data = pd.read_csv("LLM_ABM_Pilot.csv", index_col=0)

# Get counts by country (excluding the header row and "List of Countries")
# The first row is the header/question text, so we skip it
data_clean = data[data['country_reside'] != 'List of Countries'].iloc[1:]  # Skip first row (header)

# Calculate participant counts by country
country_counts = data_clean['country_reside'].value_counts()
print("Participant counts by country:")
print(country_counts)
print(f"\nTotal (Global): {len(data_clean)}")

# Store these for later use
participant_counts = {
    'India': int(country_counts.get('India', 0)),
    'Peru': int(country_counts.get('Peru', 0)),
    'United States of America': int(country_counts.get('United States of America', 0)),
    'Global': len(data_clean)
}
print("\nClean mapping:")
for country, count in participant_counts.items():
    print(f"  {country}: {count}")
```

```python
# Let's verify the counts more carefully
import pandas as pd
data = pd.read_csv("LLM_ABM_Pilot.csv", index_col=0)

# The first row (index label "1") is the question text
print("Question text row (index 1):")
print(data.loc[1, 'country_reside'])
print()

# Skip the question text row (iloc[0]) and get the rest
data_clean = data.iloc[1:]  # This skips index label "1"
print(f"Total rows after skipping question text: {len(data_clean)}")

# Filter out invalid entries
data_clean = data_clean[data_clean['country_reside'].notna()]
print(f"After removing NaN: {len(data_clean)}")

data_clean = data_clean[data_clean['country_reside'] != 'List of Countries']
print(f"After removing 'List of Countries': {len(data_clean)}")
print()

# Get countsI
country_counts = data_clean['country_reside'].value_counts()
print("Counts by country:")
print(country_counts)
print(f"\nTotal: {len(data_clean)}")
```
